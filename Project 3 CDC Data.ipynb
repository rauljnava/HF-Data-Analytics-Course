```python
#I will analyze how Medicated Patients correlate with Time in Hospital, select 'diabetesMed', 'time_in_hospital' and unique identifier 'encounter_id'
df2 = df[['encounter_id', 'diabetesMed', 'time_in_hospital']]
```

```python
df2
```

```python
#Slicing "Yes" value in the 'diabetesMed' column
df3 = df2[df2['diabetesMed'].str.contains('Yes')]
```

```python
#Slicing "No" value in the 'diabetesMed' column
df4 = df2[df2['diabetesMed'].str.contains('No')]
```

```python
#Group 1 analysis: Diabetes medicated patients: Slicing 14 value in the 'time_in_hospital' column and selecting 
#'encounter_id' as unique identifier
df5 = df3[df3['time_in_hospital'] == 14]
df6 = df5[['time_in_hospital', 'encounter_id']]
```

```python
#Counting total of group 1 patients with 14 days in the hospital. Storing result in a separate dataframe
#for further comparative analysis
df7 = df6.groupby(['time_in_hospital'])['encounter_id'].apply(pd.to_numeric).count()
```

```python
df7
```

```python
#Group 2 analysis: Diabetes not medicated patients: Slicing 14 value in the 'time_in_hospital' 
#Counting total of group 1 patients with 14 days in the hospital. Storing result in a separate dataframe
#for further comparative analysis
df8 = df4[df4['time_in_hospital'] == 14]
df9 = df8.groupby(['time_in_hospital'])['encounter_id'].apply(pd.to_numeric).count()
```

```python
df9
```

```python
#Create a data array to compare both group results
labels = ['With Diabetes Meds', 'Without Diabetes Meds']
my_data = [df7, df9]
arr = np.array(my_data)
d = {'With Diabetes Meds':df7, 'Without Diabetes Meds':df8}
```

```python
df10 = pd.Series(data=my_data, index=labels)
```
